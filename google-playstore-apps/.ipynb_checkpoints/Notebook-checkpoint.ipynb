### low_memory warning 
Guessing dtypes for each column is very memory demanding. Pandas tries to determine what dtype to set by analyzing the data in each column.
[source](https://stackoverflow.com/questions/24251219/pandas-read-csv-low-memory-and-dtype-options)

```python
def tipos(listColumns):
    for j in listColumns:
        print(j)
        listTypes = []
        for i in df_PlayStore[j].values:
            if not type(i) in listTypes:    
                print(i,type(i))
                listTypes.append(type(i))
        print()

        
tipos(df_PlayStore.columns)
```
